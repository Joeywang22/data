```python
#This array will be used to determine the colors of the data points in the scatter plot.
color=np.random.randint(0,10,150)
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=100, 
            c=color, 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=100, 
            c=color, 
            cmap="Reds", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
cbar=plt.colorbar()
```
