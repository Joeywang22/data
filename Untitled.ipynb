```python
#group by year caculate 'lifeexp' mean 
df.groupby('year')['lifeExp'].mean()
```

```python
df.groupby('country')['gdpPercap'].mean()
```

```python
#mutiple caculation 
df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
```

```python
#另一种表达方法
df\
    .groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']]\
    .mean()
```

```python
# 和上面的group不一样，并且按照year和continent 的字母排序
(df
    .groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']]
    .mean()
    .reset_index()
)
```
